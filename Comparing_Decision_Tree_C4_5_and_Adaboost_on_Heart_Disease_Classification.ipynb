# Modelling

sebelum modeling kita perlu melakukan scaling karena antar variable berbeda skalanya

```python
df.head()
```

```python
# assign to x and y
x = df.drop(columns='heart disease')
y = df['heart disease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, stratify=y)
```

```python
from sklearn.preprocessing import RobustScaler
rbs = RobustScaler()
rbs.fit(x_train)
x_train_scaled = rbs.transform(x_train)
x_test_scaled = rbs.transform(x_test)
```

```python
x_train_scaled.shape, x_test_scaled.shape
```

```python
x_train_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier

c45_classifier = DecisionTreeClassifier(criterion="entropy")
c45_classifier.fit(x_train_scaled, y_train)
y_pred = c45_classifier.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score


# Initialize C4.5 decision tree classifier
c45 = DecisionTreeClassifier(criterion="entropy")

# Initialize Adaboost classifier with C4.5 as the base estimator
adaboost_classifier = AdaBoostClassifier(estimator=c45)

# Train Adaboost classifier
adaboost_classifier.fit(x_train_scaled, y_train)

# Make predictions on the testing set
y_pred1 = adaboost_classifier.predict(x_test_scaled)

print(classification_report(y_test, y_pred1))
```

```python
cm = confusion_matrix(y_test, y_pred1)
sns.heatmap(cm, annot=True, fmt='g')
```
